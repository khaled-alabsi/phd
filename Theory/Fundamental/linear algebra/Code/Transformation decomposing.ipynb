```python
# using step-by-step transformation to reach the target vector
 
import numpy as np
import matplotlib.pyplot as plt

# Define the original transformation matrix
A = np.array([[4, 2], [1, 3]])

# Eigendecompose A
eigenvalues, eigenvectors = np.linalg.eig(A)
V = eigenvectors
Lambda = np.diag(eigenvalues)

# Number of steps
n_steps = 3

# Compute the n-th root of the transformation matrix
Lambda_root = np.diag(eigenvalues ** (1 / n_steps))
B = V @ Lambda_root @ np.linalg.inv(V)

# Define the initial vector and the target vector
start_vector = np.array([1, 1])  # Starting vector
target_vector = (
    A @ start_vector
)  # Final target vector after full transformation

# Transform the vector step-by-step
step_vectors = [start_vector]
v = start_vector
for _ in range(n_steps):
    v = B @ v
    step_vectors.append(v)

step_vectors = np.array(step_vectors)

# Plot the transformation steps, start vector, and target vector
plt.figure(figsize=(8, 8))
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")

# Plot the start vector
plt.quiver(
    0,
    0,
    start_vector[0],
    start_vector[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="green",
    label="Start Vector",
)

# Plot the step vectors
for i in range(len(step_vectors) - 1):
    plt.quiver(
        step_vectors[i, 0],
        step_vectors[i, 1],
        step_vectors[i + 1, 0] - step_vectors[i, 0],
        step_vectors[i + 1, 1] - step_vectors[i, 1],
        angles="xy",
        scale_units="xy",
        scale=1,
        color="blue",
        alpha=0.7,
    )

# Plot the target vector
plt.quiver(
    0,
    0,
    target_vector[0],
    target_vector[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="red",
    label="Target Vector",
)

# Add grid and legend
plt.grid(True, linestyle="--", alpha=0.6)
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.gca().set_aspect("equal", adjustable="box")
plt.title("Transformation: Start to Target with Steps")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
# using % decomposition

import numpy as np
import matplotlib.pyplot as plt

# Define the original transformation matrix
A = np.array([[4, 2], [1, 3]])

# Eigendecompose A
eigenvalues, eigenvectors = np.linalg.eig(A)
V = eigenvectors
Lambda = np.diag(eigenvalues)

# Define percentages for each step
percentages = [20, 70, 10]  # Must sum to 100%

# Compute fractional matrices for each percentage
fractional_matrices = []
for p in percentages:
    Lambda_frac = np.diag(eigenvalues ** (p / 100))
    B = V @ Lambda_frac @ np.linalg.inv(V)
    fractional_matrices.append(B)

# Initial vector
start_vector = np.array([1, 1])  # Starting vector
vectors = [start_vector]

# Apply the transformation sequentially
v = start_vector
for B in fractional_matrices:
    v = B @ v
    vectors.append(v)

vectors = np.array(vectors)

# Compute the target vector
target_vector = A @ start_vector

# Plot the transformation steps, start vector, and target vector
plt.figure(figsize=(8, 8))
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")

# Plot the start vector
plt.quiver(
    0,
    0,
    start_vector[0],
    start_vector[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="green",
    label="Start Vector",
)

# Plot the step vectors
for i in range(len(vectors) - 1):
    plt.quiver(
        vectors[i, 0],
        vectors[i, 1],
        vectors[i + 1, 0] - vectors[i, 0],
        vectors[i + 1, 1] - vectors[i, 1],
        angles="xy",
        scale_units="xy",
        scale=1,
        color="blue",
        alpha=0.7,
    )

# Plot the target vector
plt.quiver(
    0,
    0,
    target_vector[0],
    target_vector[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="red",
    label="Target Vector",
)

# Add grid and legend
plt.grid(True, linestyle="--", alpha=0.6)
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.gca().set_aspect("equal", adjustable="box")
plt.title("Transformation: Specific Percentage Steps")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
