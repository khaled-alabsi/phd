```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated SPE values: first 50 are normal (low), last 50 are faulty (high)
np.random.seed(0)
normal_spe = np.random.normal(loc=0.05, scale=0.02, size=50)
faulty_spe = np.random.normal(loc=50, scale=10, size=50)
spe_values = np.concatenate([normal_spe, faulty_spe])

x = np.arange(len(spe_values))

# Plot side-by-side: linear vs log scale
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Linear scale plot
ax[0].plot(x, spe_values)
ax[0].set_title("Linear Scale")
ax[0].set_xlabel("Sample")
ax[0].set_ylabel("SPE")

# Log scale plot
ax[1].plot(x, spe_values)
ax[1].set_yscale('log')
ax[1].set_title("Log Scale")
ax[1].set_xlabel("Sample")
ax[1].set_ylabel("SPE (log)")

plt.tight_layout()
plt.show()
```



### **Left plot: Linear Scale**

* The **y-axis** increases at a **constant rate**: 0, 10, 20, ..., 100
* The **normal SPE values** (near 0.05) look **almost flat**, barely visible near the x-axis.
* The **faulty SPE values** dominate the plot, taking up most of the y-range.
* You **cannot easily see the structure or variation** in the normal (in-control) data.

---

### **Right plot: Logarithmic Scale (`set_yscale('log')`)**

* The **y-axis increases multiplicatively**: 0.01, 0.1, 1, 10, 100, ...
* **Normal and faulty values are both visible**:

  * The low values are **stretched out**, making it easier to see variation during the normal phase.
  * The high values are **compressed**, so they dont overwhelm the scale.
* You can clearly see the **transition** from normal to faulty data.

---

### **Summary**

| Feature          | Linear Scale                   | Log Scale (`set_yscale('log')`)         |
| ---------------- | ------------------------------ | --------------------------------------- |
| Small values     | Compressed near 0              | Spread out (more readable)              |
| Large values     | Dominate the plot              | Compressed (less dominant)              |
| Detecting faults | Hard to see start of deviation | Easier to spot transition or spike      |
| Useful when      | All values in same range       | Values span several orders of magnitude |

---

### **When to use log scale**

Use `.set_yscale('log')` when your data:

* Spans **multiple orders of magnitude**
* Has **anomalies or outliers** much larger than the normal range
* Needs better **visual separation of low vs high regions**, e.g., SPE, loss curves, or exponential growth
