```python
import pandas as pd

# Create the DataFrame
data = pd.DataFrame({
    'Store A': [200, 220, 250, 230, 210],
    'Store B': [150, 180, 200, 190, 160],
    'Store C': [300, 310, 320, 300, 290],
    'Store D': [250, 260, 270, 280, 265],
    'Store E': [100, 90, 110, 120, 95]
}, index=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])

# Rank the data
ranked_sales = data.rank(axis=1)

print("Original Sales Data:\n", data)
print("\nRanked Sales Data:\n", ranked_sales)
```
